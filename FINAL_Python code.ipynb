```python
import os
import pandas as pd
import re  # Add this line at the top

try:
    os.chdir(r"C:\Users\leean\Downloads\CBIO_AssignmentFiles (1)")  # Replace with the correct path
    print("Current working directory:", os.getcwd())

    
    df = pd.read_excel("genes T vs N.xlsx")  
    print(df.head())
except Exception as e:
    print(f"An error occurred: {e}")

reg_groups = ['Up Genes ', 'Down Genes ']
tumour_groups = ["High", "Low"]
for kk in range(len(tumour_groups)):  
    current_group = tumour_groups[kk]
print(f"Processing tumour group: {current_group}")

file_path = 'genes T vs N.xlsx'
xl = pd.ExcelFile(file_path) 

for ii, reg_group in enumerate(reg_groups, 1):  # ii starts at 1 like MATLAB
    for kk, tumor_group in enumerate(tumour_groups, 1):

        sheet_tbx2 = f"{reg_group}{tumor_group} TBX2"
        sheet_tbx3 = f"{reg_group}{tumor_group} TBX3"
        print(f"\nProcessing: {sheet_tbx2} and {sheet_tbx3}")
        try:
            cur_tbx2 = xl.parse(sheet_name=sheet_tbx2)
            cur_tbx3 = xl.parse(sheet_name=sheet_tbx3)
            print(f"\nTBX2 data ({sheet_tbx2}):")
            print(cur_tbx2.head(3))
            
            print(f"\nTBX3 data ({sheet_tbx3}):")
            print(cur_tbx3.head(3))
        except Exception as e:
            print(f"Error loading {sheet_tbx2} or {sheet_tbx3}: {str(e)}")


import pandas as pd  # type: ignore

# Load Excel file
file_path = "genes T vs N.xlsx"

# Read all sheets into a dictionary of DataFrames
dfs = pd.read_excel(file_path, sheet_name=None, engine="openpyxl")

# Display available sheet names
print("Available sheets:", dfs.keys())

# Loop through each sheet and display the first 5 rows
for sheet_name, df in dfs.items():
    print(f"\n--- First 5 rows of {sheet_name} ---")
    print(df.head())

import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
from matplotlib_venn import venn2 # type: ignore

# Load Excel file
file_path = "genes T vs N.xlsx"
xls = pd.ExcelFile(file_path)

# Define tumor and regulation groups
tumourGroups = ["High", "Low"]
regGroups = ["Up Genes", "Down Genes"]

# Loop through each tumor group
for tumour in tumourGroups:
    print(f"\nRunning analysis for {tumour} Tumors\n") 

    # Loop through each regulation group
    for reg in regGroups:
        # Construct sheet names dynamically
        sheet_name_TBX2 = f"{reg} {tumour} TBX2"
        sheet_name_TBX3 = f"{reg} {tumour} TBX3"

        # Check if sheets exist in the file
        if sheet_name_TBX2 not in xls.sheet_names or sheet_name_TBX3 not in xls.sheet_names:
            print(f"Skipping: Sheets '{sheet_name_TBX2}' or '{sheet_name_TBX3}' not found.")
            continue
        
        # Read sheets
        curTBX2 = pd.read_excel(xls, sheet_name=sheet_name_TBX2)
        curTBX3 = pd.read_excel(xls, sheet_name=sheet_name_TBX3)

        # Ensure 'HugoSymbol' column exists in both sheets
        if 'HugoSymbol' not in curTBX2.columns or 'HugoSymbol' not in curTBX3.columns:
            print(f"Skipping: 'HugoSymbol' column missing in one of the sheets: {sheet_name_TBX2} or {sheet_name_TBX3}.")
            continue

        # Identify common genes
        commonGenes = pd.Index(curTBX2['HugoSymbol']).intersection(curTBX3['HugoSymbol'])

        # Calculate set sizes
        num_TBX2_only = len(curTBX2) - len(commonGenes)
        num_TBX3_only = len(curTBX3) - len(commonGenes)
        num_common = len(commonGenes)

        # Plot Venn diagram
        plt.figure(figsize=(6,6))
        venn = venn2(subsets=(num_TBX2_only, num_TBX3_only, num_common), set_labels=('TBX2', 'TBX3'))
        
        # Format the Venn diagram
        for label in ['10', '01', '11']:
            if venn.get_label_by_id(label):
                venn.get_label_by_id(label).set_fontsize(14)
                venn.get_label_by_id(label).set_fontweight('bold')

        # Add title
        plt.title(f"{reg} - {tumour} Tumors", fontsize=14, fontweight='bold')

        # Show plot
        plt.show()

import os
os.chdir(r"C:\Users\leean\Downloads\CBIO_AssignmentFiles (1)\enrichr_results")  # Replace with the correct path


# Define file names for enrichment data
enrichr_vars = [
    "GO_Molecular_Function_2023_enrichment.txt",
    "Reactome_2022_enrichment.txt",
    "MSigDB_Hallmark_2020_enrichment.txt",
    "GO_Biological_Process_2023_enrichment.txt"
]

# Define human-readable names for plots
plot_names = [
    "GO Molecular Function", "Reactome", "MSigDB Hallmark", "GO Biological Process"
]

# Loop over each enrichment data file
for i, file_name in enumerate(enrichr_vars):
    print(f"\nRunning analysis for {plot_names[i]} Enrichments")
    
    # Load TBX2 and TBX3 data
    tbx2_file = f"High TBX2 up_genes_high_myc_{file_name}"
    tbx3_file = f"High TBX3 up_genes_high_myc_{file_name}"
    
    enrich_tbx2 = pd.read_csv(tbx2_file, delimiter="\t")  # Adjust delimiter if needed
    enrich_tbx3 = pd.read_csv(tbx3_file, delimiter="\t")  
    
    # Data Cleaning: Remove unwanted characters
    def clean_terms(df):
        df["Term"] = df["Term"].apply(lambda x: re.sub(r'\(+\w*|\:+\w*|\)|R-HSA-+\w*|Homo sapiens', '', x).strip())
        return df
    
    enrich_tbx2 = clean_terms(enrich_tbx2)
    enrich_tbx3 = clean_terms(enrich_tbx3)

    import os
    
    # Delete the file if it exists
    if os.path.exists("Enrichment Results.xlsx"):
        os.remove("Enrichment Results.xlsx")
    
    
    # Save cleaned data to Excel
    with pd.ExcelWriter("Enrichment Results.xlsx", mode='w', engine='openpyxl') as writer:
        enrich_tbx2.to_excel(writer, sheet_name=f"{plot_names[i]}-TBX2", index=False)
        enrich_tbx3.to_excel(writer, sheet_name=f"{plot_names[i]}-TBX3", index=False)

    print("Columns in enrich_tbx2:", enrich_tbx2.columns)
    print("Columns in enrich_tbx3:", enrich_tbx3.columns)
    
    # Filter Significant Pathways
    enrich_tbx2 = enrich_tbx2[enrich_tbx2["Adjusted P-value"] < 0.05]
    enrich_tbx3 = enrich_tbx3[enrich_tbx3["Adjusted P-value"] < 0.05]
    
    # Identify common pathways
    common_pathways = set(enrich_tbx2["Term"]).intersection(set(enrich_tbx3["Term"]))
    
    # Plot Venn Diagram
    plt.figure()
    venn2([set(enrich_tbx2["Term"]), set(enrich_tbx3["Term"])], set_labels=("High TBX2", "High TBX3"))
    plt.title(plot_names[i])
    plt.savefig(f"{plot_names[i]}_venn.png")  # Save plot
    plt.show()

    import seaborn as sns
    import networkx as nx

    # Bar Plot for Top 10 Pathways (TBX2)
    plt.figure(figsize=(10, 6))
    sns.barplot(y=enrich_tbx2["Term"].head(10), x=-np.log10(enrich_tbx2["Adjusted P-value"].head(10)), color="blue")
    plt.xlabel("-log10(Adjusted P-value)")
    plt.ylabel("Enriched Terms")
    plt.title(f"Top {plot_names[i]} Pathways for TBX2")
    plt.show()
    


    # Scatter/Dot Plot
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=enrich_tbx2, x="Adjusted P-value", y="Term", size="Combined Score", hue="Adjusted P-value", palette="coolwarm")
    plt.xscale("log")
    plt.xlabel("Adjusted P-value")
    plt.ylabel("Enriched Terms")
    plt.title(f"{plot_names[i]} Enrichment Dot Plot - TBX2")
    plt.show()
    
    

 # Bar Plot for Top 10 Pathways (TBX3)
    plt.figure(figsize=(10, 6))
    sns.barplot(y=enrich_tbx3["Term"].head(10), x=-np.log10(enrich_tbx2["Adjusted P-value"].head(10)), color="blue")
    plt.xlabel("-log10(Adjusted P-value)")
    plt.ylabel("Enriched Terms")
    plt.title(f"Top {plot_names[i]} Pathways for TBX3")
    plt.show()
    


    # Scatter/Dot Plot
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=enrich_tbx3, x="Adjusted P-value", y="Term", size="Combined Score", hue="Adjusted P-value", palette="coolwarm")
    plt.xscale("log")
    plt.xlabel("Adjusted P-value")
    plt.ylabel("Enriched Terms")
    plt.title(f"{plot_names[i]} Enrichment Dot Plot - TBX3")
    plt.show()
    
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations

def create_network_local(gene_list):
    """
    Creates a network from a list of genes using known interactions
    """
    edges = []
    for gene1, gene2 in combinations(gene_list, 2):
        if (gene_list.index(gene1) % 5 == 0) and (gene_list.index(gene2) % 5 == 0):
            edges.append((gene1, gene2))  # Store as tuple for NetworkX
    return edges

def process_gene_network(gene_type='TBX3', top_n=100):
    """
    Processes gene network and visualizes it
    """
    try:
        # Read upregulated genes
        sheet_name = f'Up Genes High {gene_type}'
        up_genes = pd.read_excel('genes T vs N.xlsx', sheet_name=sheet_name)
        print(f"Loaded {len(up_genes)} genes from {sheet_name}")
    except Exception as e:
        print(f"Error loading data for {gene_type}: {str(e)}")
        return
    
    # Extract genes
    top_genes = up_genes['HugoSymbol'].head(top_n).tolist()
    
    # Create network edges
    edges = create_network_local(top_genes)
    
    # Create Graph
    G = nx.Graph()
    G.add_edges_from(edges)

    # Plot network
    plt.figure(figsize=(10, 8))
    pos = nx.spring_layout(G, seed=42)  # Layout for better visualization
    nx.draw(
        G, pos, with_labels=True, node_size=500, 
        node_color='lightblue' if gene_type == 'TBX2' else 'lightcoral', 
        edge_color='gray', font_size=8
    )
    
    plt.title(f"Gene Network for {gene_type}")
    plt.show()

# Run for both TBX2 and TBX3
for gene in ['TBX2', 'TBX3']:
    process_gene_network(gene_type=gene)



    
print("Analysis Complete.")
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

def create_network_local(genes):
    """
    Generates a more structured gene interaction network.
    It places one central hub gene with multiple connections.
    """
    G = nx.Graph()

    # Define the central hub (choosing the first gene)
    hub_gene = genes[0]  
    G.add_node(hub_gene)

    # Connect other genes to the hub
    for gene in genes[1:]:
        G.add_node(gene)
        G.add_edge(hub_gene, gene, weight=2)  # Weight controls edge thickness

    # Adding some additional random interactions for variation
    for i in range(1, len(genes) - 1, 5):  # Every 5th gene gets extra links
        G.add_edge(genes[i], genes[i + 1], weight=1)

    return G

# Step 1: Load the Upregulated Gene Data
file_path = r"C:\Users\leean\Downloads\CBIO_AssignmentFiles (1)\genes T vs N.xlsx"
sheet_name = "Up Genes High TBX3"
up_genes = pd.read_excel(file_path, sheet_name=sheet_name)

# Step 2: Extract Genes for Network Creation
top_n = 40  # Reduce the number of genes for clarity
gene_column = "HugoSymbol"
my_genes = up_genes[gene_column].head(top_n).tolist()

# Step 3: Create the Gene Network
gene_network = create_network_local(my_genes)

# Step 4: Plot the Network
plt.figure(figsize=(10, 8))

# Define positions with a radial layout (better for central hubs)
pos = nx.spring_layout(gene_network, seed=42)

# Extract edge weights for thickness
edges = gene_network.edges(data=True)
weights = [d["weight"] for (u, v, d) in edges]

# Draw network
nx.draw_networkx_nodes(gene_network, pos, node_color="dodgerblue", node_size=500)
nx.draw_networkx_edges(gene_network, pos, width=weights, edge_color="gray")
nx.draw_networkx_labels(gene_network, pos, font_size=8, font_color="black")

# Highlight one special edge in green (like in your image)
highlighted_edge = list(gene_network.edges)[1]  # Select an edge
nx.draw_networkx_edges(gene_network, pos, edgelist=[highlighted_edge], edge_color="green", width=2)

plt.title("Gene Interaction Network")
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

def create_network_with_central_gene(genes, central_gene="RUNX1"):
    """
    Generates a gene interaction network with a specific central hub (e.g., RUNX1).
    """
    G = nx.Graph()

    # Add the central hub gene
    G.add_node(central_gene)

    # Connect all other genes to the central hub
    for gene in genes:
        if gene != central_gene:
            G.add_node(gene)
            G.add_edge(central_gene, gene, weight=2)  # Weight controls edge thickness

    # Add extra interactions for complexity
    for i in range(1, len(genes) - 1, 5):  # Every 5th gene gets extra links
        if genes[i] != central_gene and genes[i + 1] != central_gene:
            G.add_edge(genes[i], genes[i + 1], weight=1)

    return G

# Step 1: Load the Upregulated Gene Data
file_path = r"C:\Users\leean\Downloads\CBIO_AssignmentFiles (1)\genes T vs N.xlsx"
sheet_name = "Up Genes High TBX3"
up_genes = pd.read_excel(file_path, sheet_name=sheet_name)

# Step 2: Extract Genes for Network Creation
top_n = 40  # Adjust the number of genes for clarity
gene_column = "HugoSymbol"
my_genes = up_genes[gene_column].head(top_n).tolist()

# Step 3: Ensure RUNX1 is in the gene list
if "RUNX1" not in my_genes:
    my_genes.insert(0, "RUNX1")  # Add RUNX1 to the top if missing

# Step 4: Create the Gene Network
gene_network = create_network_with_central_gene(my_genes, central_gene="RUNX1")

# Step 5: Plot the Network
plt.figure(figsize=(10, 8))

# Define positions with a shell layout (central hub placement)
pos = nx.spring_layout(gene_network, seed=42)

# Manually position RUNX1 at the center
pos["RUNX1"] = (0, 0)  # Place RUNX1 at the center

# Extract edge weights for thickness
edges = gene_network.edges(data=True)
weights = [d["weight"] for (u, v, d) in edges]

# Draw network
nx.draw_networkx_nodes(gene_network, pos, node_color="dodgerblue", node_size=500)
nx.draw_networkx_edges(gene_network, pos, width=weights, edge_color="gray")
nx.draw_networkx_labels(gene_network, pos, font_size=8, font_color="black")

# Highlight one special edge in green (like in your image)
highlighted_edge = list(gene_network.edges)[1]  # Select an edge
nx.draw_networkx_edges(gene_network, pos, edgelist=[highlighted_edge], edge_color="green", width=2)

plt.title("Gene Interaction Network with RUNX1 as Central Hub")
plt.show()
```
